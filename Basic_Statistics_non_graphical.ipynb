### Central Tendency Measures

```python
#to compute mean of a column in the data frame
ene['gdp_per_capita'].mean()
```

```python
#to compute median of a column in the data frame
ene['gdp_per_capita'].median()
```

```python
#to compute mode of a column in the data frame
ene['gdp_per_capita'].mode()
```

```python
#to compute the quantile of a column in the data frame. check for quantile (0), quantile(1), quantile(.25),quantile(.75)
ene['gdp_per_capita'].quantile()
```

### Measures of dispersion

```python
#to compute variance of a column in the data frame
ene['Electricity from fossil fuels (TWh)'].var()
```

```python
#to compute standard deviation of a column in the data frame
ene['Electricity from fossil fuels (TWh)'].std()
```

```python
#correlation -to check whether linear relationship exists between 2 variables
ene['Electricity from fossil fuels (TWh)'].corr(ene['gdp_growth'])
```

```python
ene['Latitude'].value_counts()
```

```python
pd.value_counts(ene['Longitude'])
```

```python
#To create frequency table
pd.crosstab(index=ene['Longitude'],columns='count',dropna=True)
```

```python
#To create two-way table
pd.crosstab(index=ene['Electricity from fossil fuels (TWh)'],columns=ene['Latitude'],dropna=True)
```

```python
#Two-way table -Joint probability
pd.crosstab(index=ene['Electricity from fossil fuels (TWh)'],columns=ene['Longitude'],normalize=True,dropna=True)
```

```python
#Two-way table -Margin probability
pd.crosstab(index=ene['Electricity from fossil fuels (TWh)'],columns=ene['Longitude'],normalize=True,margins=True,dropna=True)
```

```python
#Two-way table -conditional probability
pd.crosstab(index=ene['Electricity from fossil fuels (TWh)'],columns=ene['Longitude'],normalize='index',margins=True,dropna=True)
```

```python
#Two-way table -conditional probability
pd.crosstab(index=ene['Electricity from fossil fuels (TWh)'],columns=ene['Longitude'],normalize='columns',margins=True,dropna=True)
```
