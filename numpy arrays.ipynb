19)Create a 2X2 matrix with values ranging from 1 to 4 and find its determinant, rank and inverse of
the matrix


```python
matrix_2x2=np.array([[1,2],[3,4]])
determinant=np.linalg.det(matrix_2x2)
rank=np.linalg.matrix_rank(matrix_2x2)
inverse_matrix=np.linalg.inv(matrix_2x2)
print("Determinant:", determinant)
print("Rank:",rank)
print("Inverse:\n",inverse_matrix)
```

20)Choose an array to show the difference between assigning and copying another array.

```python
original_array=np.array([1,2,3])
assigned_array=original_array
copied_array=np.copy(original_array)
original_array[0]=10
print("Original array:",original_array)
print("Assigned array:",assigned_array)
print("Copied array:",copied_array)
```
