```python
#importing library
import seaborn as sns
```

```python
ax=sns.scatterplot(data=ene,x='Access to electricity (% of population)',y='Access to clean fuels for cooking')
ax.set_title('Access to electricity (% of population) Vs. Access to clean fuels for cooking')
```

```python
#scatter plot - more than 2 variables
ax=sns.scatterplot(data=ene,x='Access to clean fuels for cooking',y='Access to electricity (% of population)', hue='Year')
ax.set_title('Access to clean fuels for cooking vs Access to electricity (% of population)')
```

```python
#scatter plot - more than 2 variables
ax=sns.scatterplot(data=ene,x='Access to clean fuels for cooking',y='Access to electricity (% of population)', hue='Year',style='Year')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title('Access to clean fuels for cooking vs Access to electricity (% of population)')
```

```python
#scatter plot - more than 2 variables
ax=sns.scatterplot(data=ene,x='Access to clean fuels for cooking',y='Access to electricity (% of population)', hue='Year',size='gdp_growth')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title('Access to clean fuels for cooking vs Access to electricity (% of population)')
```

```python
#scatter plot
sns.relplot(data=ene,x='Access to clean fuels for cooking',y='Access to electricity (% of population)')
```

```python
#scatter plot
sns.relplot(data=ene,x='Access to clean fuels for cooking',y='Access to electricity (% of population)', hue='Year', style='Year', size='gdp_growth',col='Year')
```

```python
ene['gdp_per_capita'].value_counts()
```

```python

ax=sns.countplot(y=ene['Year'].sort_values(ascending=False)) #horizontal bar
ax.set_title("frequency of Fuel Types")
```

```python

sns.barplot(data=ene,y='gdp_growth',x='Year')
```

```python
#histogram
sns.histplot(ene['gdp_per_capita'])
```

```python
#density plot
sns.kdeplot(ene['Value_co2_emissions_kt_by_country'])
```

```python
#box plot
sns.boxplot(data=ene,x='Value_co2_emissions_kt_by_country')  #visualizing summary statistics
```

```python
sns.boxplot(data=ene, x='Value_co2_emissions_kt_by_country', y='gdp_growth', hue='Year')
```
